```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y = np.sin(x)
```

```python
y_exp = np.exp(x)
```

```python
plt.plot(x, y)
plt.plot(x, y_exp)
plt.show()
```
